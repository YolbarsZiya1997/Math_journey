### Newton's Law of Universal Gravitation

$$F = \frac{Gm_{1}m_{2}}{r^2}$$

#### Let's take the masses of two bodes: the mass of the first body(m1) is 0.5 kg, and the mass of the second body(m2) is 1.5 kg. 

$$m_{1} = 0.5 kg$$
$$m_{2} = 1.5 kg$$
$$N = 6.674\times 10^{-11}Nm^2kg^-2$$ 

#### Now we're ready to calculate the gravitational force between these two bodies at 19 different distances:100m, 150m, 200m, 250m, 300m, and so on up through 1000 m. The following program performs these calculations adn also draws the graph

```python
"""
The relationship between gravitational force and distance between two bodies
"""

import matplotlib.pyplot as plt

# Draw the graph
def draw_graph(x, y):
    plt.plot(x, y, marker='o')
    plt.xlabel('Distance in meters')
    
    plt.ylabel('Gravitational force in newtons')
    plt.title('Gravitational force and distance')
    plt.show()
    
def generate_F_r():
    # Generate values for r 
    r = range(100, 1001, 50)
    # Empty list to store the calculated values of F
    F = []
    
    # Constant, G
    G = 6.674*(10**-11)
    # Two masses
    m1 = 0.5
    m2 = 1.5
    
    # Calculate force and add it to the list, F
    for dist in r:
        force = G*(m1*m2) / (dist**2)
        F.append(force)
        
    # Call the draw_graph function
    draw_graph(r, F)

if __name__ == '__main__':
    generate_F_r()
```

### Projectile Motion

<img src="Screenshot_46.png"/>

#### The graph shows us a type of motion that is called a $projectile$ motion.

#### Initially the object is at velocity u with $\vec{u} = ucos(\theta)\hat{\boldsymbol\imath} + usin(\theta)\hat{\boldsymbol\jmath}$ The Vx component doesn't change by time (no force is acting on it) but the Vy component change in time due to the gravitational force

#### And at some point during the motion the velocity is represented by $\vec{v} = v_x\hat{\boldsymbol{\imath}} + v_y\hat{\boldsymbol{\jmath}}$ and as we discussed $v_y$ changes in time and it is

$$v_y = usin(\theta) - gt$$ 

#### By taking the integral of this expression we get

$$S_y = u(sin\theta)t - \frac{1}{2} gt^2$$

#### the vertical distance travel by the object is given by this formula

#### to be able to write the program we must know how long does the ball stay airborn, and the ball reaches its maximum hight at $v_y = 0$ thus we can say

$$v_y = usin(\theta) - gt$$  
$$t = \frac{usin\theta}{g}$$

#### let's call this time length t_peak(from 0 to y_max) the total time for the object to be air born would be
$$t_{flight} = 2t_{peak} = 2\frac{usin\theta}{g}$$

#### Let's take a ball that's thrown with an initial velocity (u) of 5m/s at an angle ($\theta$) of 45 degrees. To calculate the total time of flight, we substitute u = 5, $\theta$ = 45 and g = 9.8 into the equation above and we would have


#### t_flight = 0.72 seconds, in the code we gonna write we calculate the coordinates every 0.001 seconds

### Generating Equally Spaced Floating Point Numbers

#### There is no function that would allow us to create a list of the numbers from 0 to 0.72 with two consecutive numbers separated by 0.001. We can use a while loop as follows to create our own function for this:

```python
"""
Generate equally spaced floating point
numbers between two given values
"""

def frange(start, final, increment): # (frange() ("Floating point" range))
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + increment
        
    return numbers
```

### Drawing the Trajectory

```python
"""
Draw the trajectory of a body in projectile motion
"""

from matplotlib import pyplot as plt
import math

def draw_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('x-coordinates')
    plt.ylabel('y-coordinates')
    plt.title('Projectile motion of a ball')
    
def frange(start, final, increment):
    
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + increment
        
    return numbers

def draw_trajectory(u, theta):
    theta = math.radians(theta)
    g = 9.8
    
    # Time of flight 
    t_fight = 2*u*math.sin(theta)/g
    # Find the time intervals
    intervals = frange(0, t_fight, 0.001)
    # List of x and y coordinates
    x = []
    y = []
    for t in intervals:
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t - 0.5*g*t*t)
        
    draw_graph(x, y)
    
if __name__ == '__main__':
    try:
        u = float(input('Enter the initial velocity (m/s): '))
        theta = float(input('Enter the angle of projection(degrees): '))
    except ValueError:
        print('You entered an invalid input')
    else:
        draw_trajectory(u, theta)
        plt.show()
```

```python
if __name__ == '__main__':
    # List of three different initial velocities
    u_list = [20, 40, 60]
    theta = 45
    for u in u_list:
        draw_trajectory(u, theta)
    # Add a legend an show the graph
    plt.legend(['20', '40', '60'])
    plt.show()
```
