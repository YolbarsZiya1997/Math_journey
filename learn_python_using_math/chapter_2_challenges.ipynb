### Challenge 1 : How Does the Temperature Vary During Vary During the Day 

#### If you enter a search term like "New York weather" in Google's search engine, you'll see, among other things, a graph showing the temperature at different times of the present day. Your task here is to re-create such graph.

```python
"""
istanbul_forcast_basic.py
"""

import matplotlib.pyplot as plt

def plot_forcast():
    time_of_day = ['4 am', '7 am', '10 am', '1 pm', '4 pm', ' 7pm', '10 pm']
    temperature = [71, 70, 74, 80, 82, 81, 76]
    time_interval = range(1, len(time_of_day)+1)
    
    plt.plot(time_interval, temperature,'o-')
    plt.xticks(time_interval, time_of_day) # useful function to remember
    plt.show()
    
if __name__ == '__main__':
    plot_forcast()
```

### Challenge No. 2: Exploring a Quadratic Function Visually 

#### In chaper 1, you learned how to find the root of a quadratic equation such as $$x^2 +2x + 1 = 0$$

#### We can turn this equation into a function by writing it as $$ y = x^2 + 2x +1$$ and plot it

```python
"""
Quadratic function calculator
"""
import matplotlib.pyplot as plt

# as always draw the graph

def draw_graph(x, y):
    plt.plot(x, y)
    plt.title('Graph of the function')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.show()
    
def quad():
    x_values = range(-100, 100, 20)
    y_values = []
    for x in x_values:
            y_values.append(x**2 + 2*x + 1)
        
    draw_graph(x_values, y_values)
    
if __name__ == '__main__':
    quad()
```
