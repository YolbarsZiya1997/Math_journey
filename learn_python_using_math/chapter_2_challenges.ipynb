### Challenge 1 : How Does the Temperature Vary During Vary During the Day 

#### If you enter a search term like "New York weather" in Google's search engine, you'll see, among other things, a graph showing the temperature at different times of the present day. Your task here is to re-create such graph.

```python
"""
istanbul_forcast_basic.py
"""

import matplotlib.pyplot as plt

def plot_forcast():
    time_of_day = ['4 am', '7 am', '10 am', '1 pm', '4 pm', ' 7pm', '10 pm']
    temperature = [71, 70, 74, 80, 82, 81, 76]
    time_interval = range(1, len(time_of_day)+1)
    
    plt.plot(time_interval, temperature,'o-')
    plt.xticks(time_interval, time_of_day) # useful function to remember
    plt.show()
    
if __name__ == '__main__':
    plot_forcast()
```

### Challenge No. 2: Exploring a Quadratic Function Visually 

#### In chaper 1, you learned how to find the root of a quadratic equation such as $$x^2 +2x + 1 = 0$$

#### We can turn this equation into a function by writing it as $$ y = x^2 + 2x +1$$ and plot it

```python
"""
Quadratic function calculator
"""
import matplotlib.pyplot as plt

# as always draw the graph

def draw_graph(x, y):
    plt.plot(x, y)
    plt.title('Graph of the function')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.show()
    
def quad():
    x_values = range(-100, 100, 20)
    y_values = []
    for x in x_values:
            y_values.append(x**2 + 2*x + 1)
        
    draw_graph(x_values, y_values)
    
if __name__ == '__main__':
    quad()
```

### Challenge No.3: Enhanced Projectile Trajectory Comparision Program

#### Your challenge here is to enhance the trajectory comparison program in a few ways. First, your program should print the time of flight, maximum horizontal distance, and maximum vertical distance traveled for each of the velocity and angle of projection combnations

```python
"""
Compare the projectile motion of a body thrown with various combinations of initial
velocity and angle of projection
"""
import matplotlib.pyplot as plt
import math 

g = 9.8

def draw_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('x-coordinate')
    plt.ylabel('y-coordinate')
    plt.title('Projectile Motion of a ball')
    
def frange(start, final, increment):
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + increment
        
    return numbers 

def trajectory(u, theta, t_flight):
    # list of x and y coordinates
    x = []
    y = []
    intervals = frange(0, t_flight, 0.001) 
    for t in intervals:
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t - 0.5*g*t**2) 
    
    # Create the graph
    draw_graph(x, y)
    
if __name__ == '__main__':
    num_trajectories = int(input('How many trajectories? '))
    
    velocities = []
    angles = []
    for i in range(1, num_trajectories + 1):
        v = input('Enter the initial velocity for trajectory {0} (m/s): '.format(i)) 
        theta = input('Enter the angeles of projection for trajectory {0} (degrees): '.format(i))
        velocities.append(float(v))
        angles.append(math.radians(float(theta)))
        
    for i in range(num_trajectories):
        # Calculate time of flight, maximum horizontal distance and 
        # maximum vertical distance 
        t_fight = 2*velocities[i]*math.sin(angles[i]) / g
        S_x = velocities[i]*math.cos(angles[i])*t_fight
        S_y = velocities[i]*math.sin(angles[i]) - 0.5*g*(t_fight/2)**2
        print('Initial velocity: {0} Angle of projection: {1}'.format(velocities[i], math.degrees(angles[i])))
        print('T: {0} S_x: {1} S_y: {2}'.format(t_fight, S_x, S_y))
        print()
        trajectory(velocities[i], angles[i], t_fight) 
        
    # Add a legend and show the graph
    legends = []
    for i in range(0, num_trajectories):
        legends.append('{0} - {1}'.format(velocities[i], math.degrees(angles[i])))
    plt.legend(legends)
    plt.show()
```
